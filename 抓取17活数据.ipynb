```python
#题目：http://www.17huo.com/list-man-0-50011123-0--2m0.html，抓取商品名称，价格和商品详情链接
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
UA = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'
HEADER = {'user-agent':UA}
url = 'http://www.17huo.com/list-man-0-50011123-0--2m0.html'
names = []
prices = []
item_infos = []
```

```python
while url:
    resp = requests.get(url,HEADER)
    bs = BeautifulSoup(resp.text,'html5lib')
    items = bs.select('ul.item > li')
    for item in items:
        name = item.select('p.item_title')[0].text.strip()
        price = item.select('span.txt_nowprice')[0].text.strip()[1:]
        item_info = item.select('a')[0].get('href').strip()
        
        names.append(name)
        prices.append(price)
        item_infos.append(item_info)
        
    if bs.select('font.page_down'):
        url = bs.select('div.pagem.product_list_pager > a')[-2].get('href').strip()
    else:
        url = None
```

```python
dic = {'商品名称':names,'商品价格':prices,'商品url详情':item_infos}
```

```python
import pandas as pd
dic_pd = pd.DataFrame(dic)
```
